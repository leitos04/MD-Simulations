```python
from ovito.io import *
from ovito.modifiers import CoordinationAnalysisModifier
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

# ◆ Pure Systems

## ▻ Radial Distribution Function

```python
rdfl= pd.read_csv('density-575-20x20/pure_liquid_at_300K/rdf-pure_liquid.txt', sep=" ", skiprows=2, header=None)
rdfc= pd.read_csv('pure_crystal/rdf-pure_crystal.txt', sep=" ", skiprows=2, header=None)

xl=rdfl[0]
yl=rdfl[1]

xc=rdfc[0]
yc=rdfc[1]

fig, ax = plt.subplots(1,2, figsize=(20,6), dpi=100, sharex=True, sharey=False)
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:pink', 'tab:olive', 'tab:brown' ]

ax[0].plot(xl,yl, color='tab:blue')
ax[0].set_title('Pure Liquid')
ax[1].plot(xc,yc, color='tab:red')
ax[1].set_title('Pure Solid')

plt.suptitle('RDF of pure systems', y=0.98, size=16)

# plt.savefig('results/rdf-analysis/rdf_pure_systems.png')
```

```python
q6r_l, lq6i_l = np.genfromtxt('density-575-20x20/pure_liquid_at_300K/dump-q.lammpstrj',skip_header=9, unpack=True, usecols=(5,6))
q6r_c, q6i_c = np.genfromtxt('pure_crystal/dump-q.lammpstrj',skip_header=9, unpack=True, usecols=(5,6))

fig, ax = plt.subplots(1,2, figsize=(20,6), dpi=100, sharex=True, sharey=False)
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:pink', 'tab:olive', 'tab:brown' ]

ax[0].hist(q6r_l, alpha=0.5, bins=50, density=False, stacked=True, color='tab:blue')
ax[0].set_title('q6 histogram of pure liquid system')
ax[1].hist(q6r_c, alpha=0.5, bins=50, density=False, stacked=True, color='tab:red')
ax[1].set_title('q6 histogram of pure crystal system')

plt.suptitle('Hexatic order parameter of pure systems', y=1.05, size=16)

# plt.savefig('plot-results/bo-analysis/histogram-pure_systems.png')
```

# ◆ Seed of 55 atoms (different lattice spacings)

## ▻ Radial Distribution Function

```python
scales=['0.4','0.5','0.6','0.7','0.8','0.9']
mydata55 = []

for i in scales:
    myfile55= pd.read_csv(f'density-575-20x20/hex-{i}/55-atoms/rdf-55-atoms.txt', sep=" ", skiprows=2, header=None)
    mydata55.append(myfile55)
    
x55=[]
y55=[]

for i in range(6):
    xx55=mydata55[i][0]
    yy55=mydata55[i][1]
    x55.append(xx55)
    y55.append(yy55)
    

for i in range(6):
    plt.plot(x55[i], y55[i], label=f'hex-{scales[i]}')
    
plt.title('RDF of crystal seed of 55 atoms')
plt.legend()    

# plt.savefig('plot-results/rdf-analysis/1025_joined_rdf_55atoms.png')
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20,8), dpi=100, sharex=True, sharey=False)
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:olive' ]

v=3
for i in range(6):
    if i<=2:
        ax[0,i].plot(x55[i], y55[i], color=colors[i])
        ax[0,i].set_title(f'hex-{scales[i]}')
    else:
        ii=i-v
        ax[1,ii].plot(x55[i], y55[i], color=colors[i])
        ax[1,ii].set_title(f'hex-{scales[i]}')
plt.suptitle('RDF of crystal seed of 55 atoms', y=0.98, size=16)

# plt.savefig('plot-results/rdf-analysis/1025_rdf_55atoms.png')
```

# ◆ RDF Analysis

## ▻ RDF vs Lattice Spacing 

```python
xlat=[1.70,1.52,1.38,1.29,1.20,1.13]
# -------------------------------------- 55 atoms --------------------------------------
mylist55 = []
for i in range(6):
    maxvalues55=max(y55[i][120:140])
    mylist55.append(maxvalues55)

# -------------------------------------- 109 atoms --------------------------------------
mylist109 = []
for i in range(6):
    maxvalues109=max(y109[i][120:140])
    mylist109.append(maxvalues109)
    
# -------------------------------------- 163 atoms --------------------------------------
mylist163= []
for i in range(6):
    maxvalues163=max(y163[i][120:140])
    mylist163.append(maxvalues163)
```

```python
mylist_rdfatoms=[mylist55, mylist109, mylist163]
atoms=['55','109','163'] # for the names

fig, ax = plt.subplots(1, 3, figsize=(20,4), dpi=100, sharex=True, sharey=False)
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:pink', 'tab:olive', 'tab:brown' ]

for i in range(3):
    ax[i].plot(xlat,mylist_rdfatoms[i],'o', color=colors[i])
    ax[i].set_title(f'{atoms[i]} atoms')

plt.suptitle('RDF vs Lattice Spacing', y=1.05, size=16)

# plt.savefig('plot-results/rdf-analysis/1025_rdf-vs-lattice.png')
```
