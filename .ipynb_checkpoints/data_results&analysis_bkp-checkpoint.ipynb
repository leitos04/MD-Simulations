# ◆ Seed of 55 atoms (different lattice spacings)

## ▻ Radial Distribution Function

```python
scales=['0.4','0.5','0.6','0.7','0.8','0.9']
mydata55 = []

for i in scales:
    myfile55= pd.read_csv(f'hex-{i}/55-atoms/rdf-55-atoms.txt', sep=" ", skiprows=2, header=None)
    mydata55.append(myfile55)
    
x55=[]
y55=[]

for i in range(6):
    xx55=mydata55[i][0]
    yy55=mydata55[i][1]
    x55.append(xx55)
    y55.append(yy55)
    

for i in range(6):
    plt.plot(x55[i], y55[i], label=f'hex-{scales[i]}')
    
plt.title('RDF of crystal seed of 55 atoms')
plt.legend()    

# plt.savefig('plot-results/rdf-analysis/rdf_55atoms_joined.png')
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3, figsize=(20,8))
fig.suptitle('RDF of crystal seed of 55 atoms')

ax1.plot(x55[0],y55[0], 'tab:blue')
ax1.set_title('hex-0.4')
ax2.plot(x55[1],y55[1], 'tab:orange')
ax2.set_title('hex-0.5')
ax3.plot(x55[2],y55[2], 'tab:green')
ax3.set_title('hex-0.6')
ax4.plot(x55[3],y55[3], 'tab:red')
ax4.set_title('hex-0.7')
ax5.plot(x55[4],y55[4], 'tab:purple')
ax5.set_title('hex-0.8')
ax6.plot(x55[5],y55[5], 'tab:brown')
ax6.set_title('hex-0.9')

# plt.savefig('plot-results/rdf-analysis/rdf_55atoms.png')
```

##  ▻ Bond Orientational Order

```python
scales=['0.4','0.5','0.6','0.7','0.8','0.9']
mydataq55 = []

for i in scales:
    myfileq55= numpy.genfromtxt(f'hex-{i}/55-atoms/dump-q.lammpstrj',skip_header=9, unpack=True, usecols=(5,6))
    mydataq55.append(myfileq55)
    
xq55=[]
yq55=[]

for i in range(6):
    xxq55=mydataq55[i][0]
    yyq55=mydataq55[i][1]
    xq55.append(xxq55)
    yq55.append(yyq55)
  
# for i in range(6):
#     plt.plot(xq55[i], 'o', label=f'hex-{scales[i]}')
    
# plt.title('Crystal seed of 55 atoms')
# plt.legend()   
```
